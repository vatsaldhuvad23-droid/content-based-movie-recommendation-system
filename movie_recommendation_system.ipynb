```python
# IMPORT LIBRARY
import numpy as np
import pandas as pd
import difflib # closest search result if spelling not match we use get_close_matches()
from sklearn.feature_extraction.text import TfidfVectorizer  # Converts text data into a matrix of numbers based on word importance
from sklearn.metrics.pairwise import cosine_similarity  # Calculates similarity scores between movies based on the angle of their text vectors
```

```python
# Data Loading
movies_data = pd.read_csv('movies.csv')
movies_data.head()
```

```python
movies_data.shape # number of rows and columns
```

```python
selected_features = ['genres','keywords','tagline','cast','director'] # selecting the relevant features for recommendation
print(selected_features)
```

```python
print(movies_data[selected_features].isnull().sum()) # check null value 
movies_data[selected_features] = movies_data[selected_features].fillna('') #fill null value with empty string
```

```python
# combining all the 5 selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
print(combined_features)
```

```python
# converting the text data to feature vectors (numbers)
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)#feature_vectors into the cosine_similarity function to generate a massive grid of scores comparing every single movie to one another.
print(similarity)
print(f'Similarity Shape: {similarity.shape}') # check rows and columns shape
```

```python
# GETTING THE MOVIE NAME FROM THE USER
movie_name = input("Enter your favourite movie name: ")
```

```python
list_of_all_titles = movies_data['title'].tolist()# create list of all the movie names
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)# finding closest match of movie name with user input
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]# finding index of movie with title
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie])) # getting a list of similar movies
print(similarity_score)
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

```python
#Movie Recommendation Sytem
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
